```python
import pandas as pd 

df = pd.read_csv('data.csv')
```

```python
df.sample(10)
```

```python
# check 'Name' column for any missing value 
print(df['Name'].isnull().sum())
```

```python
# extract the title from Full name 
def get_title(name): 
    if '.' in name: 
        first_name = name.split(',')[1] # extract the first name 
        title = first_name.split('.')[0] # extract the title 
        title = title.strip() # removes whitespaces if any 
        
        return title 
    else: 
        return 'No Title'

    

# demostrating use of get_title function  

# generate a set containing all titles 
# titles = set([x for x in df.Name.map(lambda x: get_title(x))])
#print(titles)




# Convert random titles to standard shorter titles 
def shorter_titles(x): 
    title = x['Title']
    if title in ['Capt', 'Col', 'Major']: 
        return 'Officer'
    elif title in ['Jonkheer', 'Don', 'the Countess', 'Lady', 'Sir']: 
        return 'Royalty'
    elif title == 'Mme': 
        return 'Mrs'
    elif title in ['Mlle', 'Ms']: 
        return 'Miss'
    else: 
        return title 
    


# finally create a column in dataframe called 'Title'
df['Title'] = df['Name'].map(lambda x: get_title(x))

# df['Title'].value_counts() # display the output 

# transform the 'Title' column values into shorter titles 
# by applying 'shorter_titles()' function to all the values for 'Title' column  
df['Title'] = df.apply(shorter_titles, axis=1)

df['Title'].value_counts() # display the output 


# delete 'Name' column if needed 
df.drop('Name', axis=1, inplace=True)

df.sample(10)
```
